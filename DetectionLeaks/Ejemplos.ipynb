# Detection Leaks - ejemplo de carga del dataset

Este notebook documenta el punto de entrada de datos para el challenge de deteccion de fugas. Usa `DatasetExample.xlsx`, ubicado en esta misma carpeta, y lo carga en estructuras de Python con `pandas`.

## Archivos de la carpeta

- `Challege_DetectionLeaks.pdf`: enunciado del challenge.
- `DatasetExample.xlsx`: dataset de ejemplo para explorar presiones nominales y mediciones asociadas a posiciones de fuga.
- `Ejemplos.ipynb`: notebook de trabajo con la carga inicial del Excel.

## Requisitos

La celda de carga necesita `pandas` y `openpyxl`. `openpyxl` ya esta instalado en este entorno. La celda busca `DatasetExample.xlsx` primero en la carpeta actual y despues junto a este notebook.


## 1. Leer `DatasetExample.xlsx` con pandas

Esta celda viene de `Master/Ejemplos.ipynb`. Lee todas las hojas del Excel y deja el resultado en `df`: un `DataFrame` si hay una sola hoja o un diccionario de `DataFrame` si hay varias.

```python
#########################################
## 1. Leer un archivo Excel con pandas ##
##########################################

import pandas as pd
from pathlib import Path

# Cell 0: leer Excel con pandas

excel_candidates = [
    Path('DatasetExample.xlsx'),
    Path('Master/DetectionLeaks/DatasetExample.xlsx'),
    Path.cwd() / 'DatasetExample.xlsx',
]

excel_file = next((candidate for candidate in excel_candidates if candidate.exists()), None)
if excel_file is None:
    raise FileNotFoundError(
        'No se encontro DatasetExample.xlsx. Ejecuta el notebook desde Master/DetectionLeaks '
        'o ajusta excel_candidates con la ruta correcta.'
    )

# Lee todas las hojas (devuelve dict de DataFrames). Para una hoja específica, usar sheet_name='NombreHoja'.
dfs = pd.read_excel(excel_file, sheet_name=None, engine='openpyxl')

# Si solo hay una hoja, obtener el DataFrame; si no, conservar el dict.
if len(dfs) == 1:
    df = next(iter(dfs.values()))
    display(df.head())
else:
    df = dfs
    print('Hojas cargadas:', list(df))
    display(next(iter(df.values())).head())
```

## 2. Inspeccion rapida del dataset

Despues de ejecutar la carga anterior, estas celdas ayudan a revisar la estructura sin asumir de antemano el numero de hojas.

```python
if isinstance(df, dict):
    print('Hojas:', list(df))
    for sheet_name, sheet_df in df.items():
        print('{}: {} filas x {} columnas'.format(sheet_name, sheet_df.shape[0], sheet_df.shape[1]))
else:
    print('Dataset: {} filas x {} columnas'.format(df.shape[0], df.shape[1]))
    display(df.head())
```

```python
active_df = next(iter(df.values())) if isinstance(df, dict) else df

summary = {
    'columns': list(active_df.columns),
    'shape': active_df.shape,
    'missing_values_by_column': active_df.isna().sum().to_dict(),
}

summary
```
